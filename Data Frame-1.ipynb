```python
#57 broadcasting operations
import pandas as pd
nba=pd.read_csv("nba.csv")
nba["Age"].add(5)
nba["Age"]+5

nba["Salary"].sub(1000000)
nba["Salary"]-(1000000)

nba["Weight"].mul(0.45)
nba["Weight"] * (0.45)

nba["Weight"].div(10000000)
nba["Weight"] / (10000000)

nba["Weight in kilograms"] = nba["Weight"].mul(0.45)
nba #it created new column in a frame
```

```python
#59 dropping NaN values from rows
nba.dropna() #it drops the rows if any column have null values
nba.dropna(how="all")# it removes the rows if all the column have null values
nba.dropna(axis=1) #it removes the columns if one number having null value
nba.dropna(subset = ["Salary","College"]) # removes null rows for the mentioned columns
```

```python
#60 Fill Null values 
import pandas as pd
nba=pd.read_csv("nba.csv")
nba["College"].fillna("No College")
nba.fillna(0)
```

```python
#61 converting data types
import pandas as pd
nba=pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"].fillna(0, inplace=True)
nba["College"].fillna("None", inplace=True)
nba.dtypes
```

```python
nba["Salary"] = nba["Salary"].astype("int")
nba["Number"] = nba["Number"].astype("int")
nba
```

```python
#62 sorting on single columns in a dataframe
import pandas as pd
nba=pd.read_csv("nba.csv")
nba.sort_values("Name", ascending=False)
nba.sort_values("Salary", ascending=False)
nba.sort_values("Salary", ascending=False, na_position = "first") #it sorts the values to the first
```

```python
#63 sorting on multiple columns in a dataframe
import pandas as pd
nba=pd.read_csv("nba.csv")
nba.sort_values(["Team","Name"], ascending = [True,False])
```

```python
#64 sorting based on sort index method
import pandas as pd
nba=pd.read_csv("nba.csv")
nba.sort_values(["Number","Salary","Name"])
nba.sort_index(ascending=False)
```

```python
# Rank Method
import pandas as pd
nba=pd.read_csv("nba.csv")
nba["Salary Rank"]=nba["Salary"].rank(ascending=False)
nba=nba.sort_values(["Salary Rank"], ascending=[False])
nba.head(10)
```
