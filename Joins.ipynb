```python
#123
import pandas as pd
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2= pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
satisfaction=pd.read_csv("Restaurant - Week 1 Satisfaction.csv")

week2
customers
foods
week1
```

```python
#124 concat append
pd.concat(objs=[week1,week2], ignore_index=False) #by default ignore index will be False
pd.concat(objs=[week1,week2], ignore_index=True)
week1.append(other=week2,ignore_index=True)
```

```python
#125 concat -2
pd.concat(objs=[week1,week2],keys=["week 1","week 2"])
```

```python
#loc method:it works on keys
import pandas as pd
df=pd.concat(objs=[week1,week2],keys=["week 1","week 2"])
df.loc[("week 1")]
df.loc[("week 1", 240)]
df.loc[("week 1", 240), ["Customer ID","Food ID"]]
```

```python
#126 inner joins-1
week1.merge(week2,how="inner",on="Customer ID", suffixes=["- week 1","- week 2"])
#data validation
#week1[week1["Customer ID"]==155]
#week2[week2["Customer ID"]==155]
```

```python
#127 inner joins-1
week1.merge(week2,how="inner",on=["Customer ID","Food ID"])
```

```python
#128 outer join
merged=week1.merge(week2,how="outer",on="Customer ID", suffixes=["- week 1","- week 2"],indicator=True)
merged
```

```python
fil=merged["_merge"].isin(["left_only","right_only"]) #it gives not matched data
merged[fil]
```

```python
#129 left join
week1.merge(foods,how="left",on="Food ID",sort=True)
```

```python
#130 left_on & right_on
week2.merge(customers, how="left",left_on="Customer ID",right_on="ID",sort=True).drop("ID",axis="columns")
```

```python
#131 Merging by indexes with left_index & right_index

foods=pd.read_csv("Restaurant - Foods.csv",index_col="Food ID")
customers=pd.read_csv("Restaurant - Customers.csv",index_col="ID")
week1.merge(customers,how="left",left_on="Customer ID",right_index=True)
week1.merge(customers,how="left",left_index=True,right_index=True)
```

```python
#132 Join method
week1.join(satisfaction)
```

```python
#133 merge method
pd.merge(week1,customers,how="left",left_on="Customer ID",right_on="ID")
```
