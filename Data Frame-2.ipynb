```python
import pandas as pd
df=pd.read_csv("employees.csv")
df.info()
```

```python
#66 memory optimization
df["Start Date"] = pd.to_datetime(df["Start Date"]) #it reduces the memory
df.head(3)
```

```python
df["Last Login Time"] = pd.to_datetime(df["Last Login Time"]) #it reduces the memory
df.head(3)

#df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"]) #either can be used 
```

```python
df["Senior Management"] = df["Senior Management"].astype("bool") #it reduces the memory
df.head(3)
```

```python
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
df.info()
#previously it was 62kb but now it is 49kb
```

```python
#67 filtering data frame on one condition
df[df["Gender"]=='Male']
##or diffferent style
fin=df["Team"]=="Finance"
df[fin]
```

```python
#68 filtering data frame on multiple condition (AND)
fil1=df["Gender"]=='Male'
fil2=df["Team"]=="Finance"
df[fil1 & fil2]
```

```python
#68,69 filtering data frame on multiple condition(OR)
fil1=df["Senior Management"]
fil2=df["Start Date"] < "1990-01-01"
fil3=df["Team"]=="Finance"
df[fil1 | fil2]
df[fil1 | fil2 & fil3]
```

```python
#70 isin method
fil=df["Team"].isin(["Legal","Product","Marketing"])
df[fil]
```

```python
#71 is null method
import pandas as pd
df=pd.read_csv("employees.csv")
fil=df["Team"].isnull()
df[fil]
```

```python
#not Null method
fil=df["Gender"].notnull()
df[fil]
```

```python
#72 between method
fil=df["Salary"].between(60000,70000)
df[fil]
```

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])#convert the start date into date format
fil=df["Start Date"].between("1991-01-01","2000-01-01")
df[fil]
```

```python
#73 Check for duplicate menthod (extarcting duplicates)
import pandas as pd
emp=pd.read_csv("employees.csv").sort_values(by="First Name")
emp[emp["First Name"].duplicated()] #gives all duplicate values
```

```python
emp[emp["First Name"].duplicated(keep = "first")
```

```python
emp[emp["First Name"].duplicated(keep = False)]
```

```python
#74 drop duplicate method
df=pd.read_csv("employees.csv").sort_values(by="First Name")
df.drop_duplicates(subset=["First Name"], keep="first") #it keeps the first value & drops the remaining duplicates
```

```python
df.drop_duplicates(subset=["First Name"], keep=False)#it removes the rows which have duplicates
```
